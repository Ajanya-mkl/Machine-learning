### missing values
1.fillna to fill missing values using different ways

2.interpolate to make a guess on missing values using interpolation

3.dropna to drop rows with missing values

```python
data = {
    'ID': range(1, 16),
    'Name': ['Alice', 'Bob', np.nan, 'David', 'Eve', 'Frank', np.nan, 'Hannah', 'Ian', 'Jack', 'Kate', 'Liam', np.nan, 'Nina', 'Oscar'],
    'Age': [25, np.nan, 22, np.nan, 30, 35, 28, np.nan, 24, 27, np.nan, 32, 29, np.nan, 26],
    'Score': [85.0, 90.5, np.nan, 88.0, 92.5, np.nan, 79.0, 95.0, np.nan, 80.0, 85.5, np.nan, 89.0, 94.0, np.nan]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
type(df["Score"][1])
```

```python
#dropping na values
df1=df.dropna()
df1
```

```python
#fillna
df2=df.fillna({"Name":"Unknown","Age":df["Age"].mean(),"Score":df["Score"].mean()})
df2
```

```python
#forward filling &backward filling
df_ffill=df.ffill() #forward filling
print(df_ffill)
df_bfill=df.bfill()
df_bfill
```

# CONCAT

```python

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Concatenate along rows
result = pd.concat([df1, df2])
print(result)
```

```python
result = pd.concat([df1, df2], axis=0)
print(result)
```

## MERGE

```python
df1,df2
```

```python


df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['B', 'C', 'D'], 'value2': [4, 5, 6]})

result = pd.merge(df1, df2, how='left', on='key')
print(result)
```

## pivot table

```python

data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03', '2023-01-04', '2023-01-04', '2023-01-05', '2023-01-05'],
    'Store': ['Store A', 'Store B', 'Store A', 'Store B', 'Store A', 'Store B', 'Store A', 'Store B', 'Store A', 'Store B'],
    'Sales': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550],
    'Revenue': [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500]
}

df = pd.DataFrame(data)
print(df)
```

```python
pivot_table = pd.pivot_table(df, values=['Sales','Revenue'], index='Date', 
                             columns='Store', aggfunc='sum')
print(pivot_table)
```
