## Bar Plot
 To compare quantities across different categories.

```python
data = {
    'Store': ['Store A', 'Store B', 'Store C'],
    'Sales': [300, 450, 250]
}

df = pd.DataFrame(data)

plt.figure(figsize=(7, 4))
plt.bar(df['Store'], df['Sales'], color='skyblue')
plt.title('Sales by Store')
plt.xlabel('Store')
plt.ylabel('Sales')
plt.show()
```

## Histogram
 To show the distribution of a numerical variable.

```python
data = {
    'Sales': [100,100,100,200, 200,300, 400,400,500,600,600, 600, 700, 800, 900,1000, 1000]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.hist(df['Sales'], bins=6, color='green', edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

## Box Plot
 To show the spread and outliers of a numerical variable.

```python
data = {
    'Revenue': [1000, 1500, 2000, 2500, 3000, 10000, 12000, 1100, 1300, 1400]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.boxplot(df['Revenue'])
plt.title('Revenue Distribution')
plt.ylabel('Revenue')
plt.show()
```

## Scatter Plot
 To show the relationship between two numerical variables.

```python
data = {
    'Sales': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'Revenue': [1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.scatter(df['Sales'], df['Revenue'], color='red')
plt.title('Sales vs Revenue')
plt.xlabel('Sales')
plt.ylabel('Revenue')
plt.show()
```

## Pie Chart
To show the proportion of categories within a whole.

```python
import pandas as pd
```

```python
data = {
    'Store': ['Store A', 'Store B', 'Store C'],
    'Sales': [300, 450, 250]
}

df = pd.DataFrame(data)

plt.figure(figsize=(4, 4))
plt.pie(df['Sales'], labels=df['Store'])
plt.title('Sales Distribution by Store')
plt.show()
```

```python
col=["green","blue","white"]
explodes=[0,0,.1]

plt.pie(df['Sales'],labels=df['Store'],colors=col,wedgeprops={'linewidth':2,"edgecolor":"black",
        "width":0.1},
        explode=explodes,
       shadow=True,startangle=90,autopct="%1.0f%%")#%23.45%%

plt.title('Sales Distribution by Store')
plt.show()
```

## Area Plot
To visualize cumulative data over time.

```python
# Sample data
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Sales': [100, 200, 300, 400, 500]
}

df = pd.DataFrame(data)

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Calculate cumulative sales 
df['Cumulative Sales'] = df['Sales'].cumsum()

# Plot
plt.figure(figsize=(10, 6))
plt.fill_between(df['Date'], df['Cumulative Sales'], color='skyblue', alpha=1)
plt.plot(df['Date'], df['Cumulative Sales'], color='blue', alpha=0.6, linewidth=2)
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales')
plt.show()
```
