```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("homeprices_new.csv")

data
```

```python
data=data.drop(["Unnamed: 0"],axis=1)
```

```python
data.isna().sum()
```

```python
data["bedrooms"]=data["bedrooms"].fillna(data.bedrooms.median())

```

```python
data.isna().sum().sum()
```

```python
plt.xlabel("Age")
plt.ylabel("Price")
plt.scatter(data.age,data.price,marker='*')
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(data.area,data.price,marker='*')
```

```python
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.scatter(data.bedrooms,data.price,marker='*')
```

```python
x=data.drop("price",axis=1)
y=data.price
```

```python
x
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python

model.intercept_

```

```python
model.predict([[3500,4,56]])

```

```python
y_pred=model.predict(x)
y_pred
```

```python
df=pd.DataFrame({"true":y,"predicted":y_pred})

```
