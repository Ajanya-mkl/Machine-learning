```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("diamonds.csv")
data
```

```python
data=data.drop("Unnamed: 0",axis=1)

```

```python
cut={'Ideal':5, 'Premium':4, 'Good':2, 'Very Good':3, 'Fair':1}
color={'E':6, 'I':2, 'J':1, 'H':3, 'F':5, 'G':4, 'D':7}
clarity={'SI2':2, 'SI1':3, 'VS1':5, 'VS2':4, 'VVS2':6, 'VVS1':7, 'I1':1, 'IF':8}
```

```python
data["cut"]=data["cut"].map(cut)
```

```python
data["color"]=data["color"].map(color)
```

```python
data["clarity"]=data["clarity"].map(clarity)
```

```python
data
```

```python
x=data.drop("price",axis=1)
y=data.price
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
scalar=MinMaxScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
x_train
```

```python
kn=KNeighborsRegressor()
kn.fit(x_train,y_train)
```

```python
kn.score(x_test,y_test)
```
